Matriks citra yang digunakan:

[148 164 174]  
[144 159 167]  
[128 147 157]

```python
import numpy as np
```

```python
f = np.array([
    [148, 164, 174],
    [144, 159, 167],
    [128, 147, 157]
], dtype=float)

print("Matriks Citra:")
print(f)
```

```python
M, N = f.shape
F = np.zeros((M, N), dtype=complex)

for u in range(M):
    for v in range(N):
        sum_val = 0
        for x in range(M):
            for y in range(N):
                exponent = -2j * np.pi * ((u * x / M) + (v * y / N))
                sum_val += f[x, y] * np.exp(exponent)
        F[u, v] = sum_val

print("Hasil DFT:")
print(F)
```

```python
magnitude = np.abs(F)

print("Magnitude Spectrum:")
print(magnitude)
```

```python
F_fft = np.fft.fft2(f)

print("DFT (FFT bawaan):")
print(F_fft)
```
