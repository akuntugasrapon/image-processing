```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("image.jpg").convert("L")
img = np.array(img)

plt.imshow(img, cmap='gray')
plt.title("Citra Asli")
plt.axis("off")
plt.show()
```

```python
kernel = np.ones((5,5)) / 25
```

```python
pad = 2
padded = np.pad(img, pad, mode='constant')
```

```python
output = np.zeros_like(img)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):

        region = padded[i:i+5, j:j+5]

        value = np.sum(region * kernel)

        output[i,j] = value
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(output, cmap='gray')
plt.title("Hasil Convolution (5x5 Mean Filter)")

plt.show()
```

```python
diff = np.abs(img - output)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,3,2)
plt.imshow(output, cmap='gray')
plt.title("Convolution 5x5")

plt.subplot(1,3,3)
plt.imshow(diff, cmap='gray')
plt.title("Difference")

plt.show()
```

```python
Image.fromarray(img).save("original.jpg")
Image.fromarray(output).save("hasil_convolution.jpg")
Image.fromarray(diff).save("difference.jpg")
```
